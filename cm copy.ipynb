```python


import numpy as np
from scipy import signal, stats


# ------------------------------------------------------------
# Filtering
# ------------------------------------------------------------

def bandpass_filter(data, sfreq, low=60.0, high=120.0, order=4):
    """Apply a Butterworth bandpass filter along the last axis."""
    nyquist = 0.5 * sfreq
    low_norm = low / nyquist
    high_norm = high / nyquist
    b, a = signal.butter(order, [low_norm, high_norm], btype="band")
    return signal.filtfilt(b, a, data, axis=-1)


# ------------------------------------------------------------
# Power computation
# ------------------------------------------------------------

def compute_power(epochs_data, sfreq=1000.0):
    """
    Compute mean squared amplitude (power) for each trial.

    Parameters
    ----------
    epochs_data : ndarray
        Shape (n_trials, n_channels, n_times). Channels of interest only.

    sfreq : float
        Sampling frequency.

    Returns
    -------
    ndarray
        Power per trial (n_trials,).
    """
    n_trials, n_channels, n_times = epochs_data.shape 

    reshaped = epochs_data.reshape(n_trials * n_channels, n_times)
    filtered = bandpass_filter(reshaped, sfreq, low=60.0, high=120.0, order=4)
    filtered = filtered.reshape(n_trials, n_channels, n_times)

    power = np.mean(filtered ** 2, axis=(1, 2))
    return power


```
