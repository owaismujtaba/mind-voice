```python
class ProgressFileWrapper:
    def __init__(self, file, pbar):
        self.file = file
        self.pbar = pbar
    def read(self, size=-1):
        data = self.file.read(size)
        self.pbar.update(len(data))
        return data
    def __getattr__(self, attr):
        return getattr(self.file, attr)
```

```python
def upload_file(local_path, remote_path):
    size = os.path.getsize(local_path)
    try:
        with open(local_path, "rb") as f, tqdm(total=size, unit="B", unit_scale=True, desc=os.path.basename(local_path)) as pbar:
            wrapped_file = ProgressFileWrapper(f, pbar)
            storage.create_file(remote_path, wrapped_file)
    except:
        print('Already Uploaded', remote_path)
```

```python
already_uploaded = []
for root, _, files in os.walk(local_dir):
    for file in files:
        flag = 0
        for uploaded in already_uploaded:
            print('Already Uploaded', file)
            if uploaded in file:
                flag=1
                break
        if flag==1:
            continue
        local_path = os.path.join(root, file)
        relative_path = os.path.relpath(local_path, local_dir).replace("\\", "/")
        remote_path = f"{remote_dir}/{relative_path}"
        upload_file(local_path, remote_path)
```
